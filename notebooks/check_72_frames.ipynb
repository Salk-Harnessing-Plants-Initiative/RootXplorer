```python
import os
import cv2
```

```python
def check_72_frames(plant_folder):
    """Check whether 72 frames in a plant folder"""
    files = [file for file in os.listdir(plant_folder) if not file.endswith('.')]
    file_count = len(files)
    return file_count
        
```

```python
def count_72(root_folder):
    wave_list = [file for file in os.listdir(root_folder) if not file.endswith('yml')]
    for wave in wave_list:
        scanner_list = [f for f in os.listdir(os.path.join(root_folder,wave)) if os.path.isdir(os.path.join(root_folder, wave,f))]
        
        for scanner in scanner_list:
            plant_list = [f for f in os.listdir(os.path.join(root_folder,wave, scanner)) if os.path.isdir(os.path.join(root_folder, wave,scanner,f))]
            
            for plant in plant_list:
                
                plant_folder = os.path.join(root_folder,wave,scanner,plant)
                file_count = check_72_frames(plant_folder)
                if file_count< 72:
                    print(f"wave: {wave}, scanner: {scanner}, plant: {plant}, frame number: {file_count)}")
                    
```

```python
def check_images(root_folder):
    problematic_images = []
    wave_list = [file for file in os.listdir(root_folder) if not file.endswith('yml')]
    for wave in wave_list:
        scanner_list = [f for f in os.listdir(os.path.join(root_folder,wave)) if os.path.isdir(os.path.join(root_folder, wave,f))]
        for scanner in scanner_list:
            plant_list = [f for f in os.listdir(os.path.join(root_folder,wave, scanner)) if os.path.isdir(os.path.join(root_folder, wave,scanner,f))]
            for plant in plant_list:
                print(f"wave: {wave}, scanner: {scanner}, plant: {plant}")
                plant_folder = os.path.join(root_folder,wave,scanner,plant)
                img_list = [file for file in os.listdir(plant_folder)]
                for img in img_list:
                    image_path = os.path.join(plant_folder,img)
                    try:
                        # Try to open the image using cv2
                        image = cv2.imread(image_path)
                        if image is None:
                            problematic_images.append(wave+'_'+scanner+'_'+plant+'_'+img)
                    except Exception as e:
                        # Handle any exceptions that occur during image opening
                        problematic_images.append((wave+'_'+scanner+'_'+plant+'_'+img, str(e)))
    return problematic_images
```
