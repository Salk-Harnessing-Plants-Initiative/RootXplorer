```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = r"Y:\Lin_Wang\Ara_cylinder\root_penetrate\col-0\Images\W1\Day_20_8-23-23_SlowScanner\AG9PQYFX84\1.png"
img = cv2.imread(img_path)
```

```python
plt.imshow(img)
```

```python
img_center = img[:,480:1630,0]
print(img_center.shape)
plt.imshow(img_center,cmap='gray')
```

```python
img_center_avgy = np.mean(img_center, axis=1)
print(img_center_avgy.shape)
```

```python
x_values = np.arange(len(img_center_avgy))

plt.scatter(img_center_avgy,x_values)
plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_center,cmap='gray')

plt.subplot(1,2,2)
plt.scatter(img_center_avgy,x_values)
plt.gca().invert_yaxis()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_center[100:250,:],cmap='gray')

plt.subplot(1,2,2)
plt.scatter(img_center_avgy[100:250],x_values[100:250])
plt.gca().invert_yaxis()
```

```python
img_center1 = img_center[:,250:350]
img_center_avgy1 = np.mean(img_center1, axis=1)
img_center2 = img_center[:,550:650]
img_center_avgy2 = np.mean(img_center2, axis=1)
img_center3 = img_center[:,850:950]
img_center_avgy3 = np.mean(img_center3, axis=1)

plt.figure(figsize=(10,12))
plt.subplot(3,2,1)
plt.imshow(img_center1,cmap='gray') #[200:250,:]

plt.subplot(3,2,2)
plt.scatter(img_center_avgy1,x_values)#
plt.gca().invert_yaxis()

plt.figure(figsize=(10,12))
plt.subplot(3,2,3)
plt.imshow(img_center2,cmap='gray')

plt.subplot(3,2,4)
plt.scatter(img_center_avgy2,x_values)
plt.gca().invert_yaxis()

plt.figure(figsize=(10,12))
plt.subplot(3,2,5)
plt.imshow(img_center3,cmap='gray')

plt.subplot(3,2,6)
plt.scatter(img_center_avgy3,x_values)
plt.gca().invert_yaxis()
```

```python
gradient_y1 = cv2.Sobel(img_center1, cv2.CV_64F, 0, 1, ksize=5)
gradient_avgy1 = np.mean(gradient_y1, axis=1)

gradient_y3 = cv2.Sobel(img_center3, cv2.CV_64F, 0, 1, ksize=5)
gradient_avgy3 = np.mean(gradient_y3, axis=1)

plt.figure(figsize=(10,8))
plt.subplot(2,3,1)
plt.imshow(img_center1[100:180,:],cmap='gray') #[200:250,:]

plt.subplot(2,3,2)
plt.imshow(gradient_y1[100:180,:],cmap='gray') #[200:250,:]

plt.subplot(2,3,3)
plt.scatter(gradient_avgy1[100:180],x_values[100:180])#
plt.gca().invert_yaxis()

plt.subplot(2,3,4)
plt.imshow(img_center3[100:180,:],cmap='gray') #[200:250,:]

plt.subplot(2,3,5)
plt.imshow(gradient_y3[100:180,:],cmap='gray') #[200:250,:]

plt.subplot(2,3,6)
plt.scatter(gradient_avgy3[100:180],x_values[100:180])#
plt.gca().invert_yaxis()
```
